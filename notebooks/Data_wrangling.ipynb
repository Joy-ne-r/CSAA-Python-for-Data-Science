```python
#Load the data
titanic_data= pd.read_excel("../data/titanic.xlsx")
join_data1= pd.read_csv("../data/joining_data1.csv")
join_data2= pd.read_csv("../data/joining_data2.csv")
union_data= pd.read_csv("../data/union_data.csv")
```

```python
#Perform a left join, joins the second data set to the first
join_data = pd.merge(left=join_data1,
                     right=join_data2,
                     how="left",
                     on="name" # common column that both has
                    )
join_data
```

```python
# Joins the first data set to the second one
join_data = pd.merge(left=join_data1,
                     right=join_data2,
                     how="right",
                     on="name" # common column that both has
                    )
join_data
```

```python
#Only common things on both sets are displayed
join_data = pd.merge(left=join_data1,
                     right=join_data2,
                     how="inner",
                     on="name" # common column that both has
                    )
join_data
```

```python
#Exverthing on both the joined sets is displayed
join_data = pd.merge(left=join_data1,
                     right=join_data2,
                     how="outer",
                     on="name" # common column that both has
                    )
join_data
```

```python
# Set plot style
plt.style.use("ggplot")
#sns.set(style='whitegrid')

# Create a figure with 3 subplots side by side
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Survival vs Pclass
sns.barplot(x='pclass', y='survived', data=titanic_data, ax=axes[0], color="red")
axes[0].set_title('Survival Rate by Passenger Class')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Survival Rate')

# Survival vs Fare (with binning for better grouping)
titanic_data['fare_band'] = pd.qcut(titanic_data['fare'], 4)
sns.barplot(x='fare_band', y='survived', data=titanic_data, ax=axes[1], color="blue")
axes[1].set_title('Survival Rate by Fare Band')
axes[1].set_xlabel('Fare Band')
axes[1].set_ylabel('Survival Rate')
axes[1].tick_params(axis='x', rotation=45)

# Survival vs Parch
sns.barplot(x='parch', y='survived', data=titanic_data, ax=axes[2], color="green")
axes[2].set_title('Survival Rate by Family Onboard')
axes[2].set_xlabel('Number of Parents/Children')
axes[2].set_ylabel('Survival Rate')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
ax = sns.barplot(x='pclass', y='survived', data=titanic_data)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')

# Add comment above the first bar
first_class_rate = titanic_data[titanic_data['pclass'] == 1]['survived'].mean()

ax.annotate('Highest rate',
            xy=(0, first_class_rate),
            xytext=(0, first_class_rate + 0.1),
            arrowprops=dict(facecolor='green', shrink=0.05),
            fontsize=12,
            color='darkgreen')


plt.tight_layout()
plt.show()
```
