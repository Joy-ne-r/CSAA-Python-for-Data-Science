```python
#Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Line plot
x = np.linspace(0,0.9,12)
y = x**2 + (4*x - 2.1)
plt.plot(x,y);
plt.xlabel("Some random numbers")
plt.ylabel("Function of some random numbers")
plt.title("Change in some random numbers")
```

```python
#Scatter plot
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,100,86,103,87,94,78,77,85,86]
plt.scatter(x,y);
plt.xlabel("Student ID'S")
plt.ylabel("WScores")
plt.title("Score of Students in National exams")
           
```

```python
#Bar Plot
categories = ["A","B","C","D"]
values = [5,7,3,8]
plt.bar(categories, values)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Number of items bought per day")
```

```python
#Histogram
import numpy as np
data = np.random.normal(0,1,100)
plt.hist(data, bins=20)
plt.ylabel("Frequency")
plt.xlabel("Values")
plt.title("The intensity of sound waves")
```

```python
plt.style.use("ggplot")
#plt.style.use("dark_background")

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

#Line Graph
x = np.linspace(0,0.9,12)
y = x**2 + (4*x - 2.1)
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title("Change in some random numbers")
axs[0, 0].set_xlabel("Some random numbers")
axs[0, 0].set_ylabel("Function of some random numbers")

#Scatter plot
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,100,86,103,87,94,78,77,85,86]
axs[0, 1].scatter(x,y,color="orange");
axs[0, 1].set_title("Students in National exams")
axs[0, 1].set_xlabel("Student ID'S")
axs[0, 1].set_ylabel("Scores")

#Bar Graph
categories = ["A","B","C","D"]
values = [5,7,3,8]
axs[1, 0].bar(categories, values, color= "pink")
axs[1, 0].set_title("Number of items bought per day")
axs[1, 0].set_xlabel("Categories")
axs[1, 0].set_ylabel("Values")

#Histogram
data = np.random.normal(0,1,100)
axs[1, 1].hist(data, bins=10,color="green")
axs[1, 1].set_title("The intensity of sound waves")
axs[1, 1].set_xlabel("Values")
axs[1, 1].set_ylabel("Frequency")

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()

# Display the plot
plt.show()
```
